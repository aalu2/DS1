```python
#Answer of 1 
# Define a Node class to create a node of linked list
class Node:
    def __init__(self, data=0, next=None): # initialize data and next for node
        self.data = data
        self.next = next

# Define a function to delete all elements with a given value in a linked list
def delete_node(head, key):
    dummy = Node(0) # create a dummy node with data set to 0 and point to head
    dummy.next = head # dummy node's next points to head
    prev = dummy # initialize prev as dummy node
    curr = head # initialize curr as head
    while curr: # loop through linked list till the end
        if curr.data == key: # if current node's data is equal to key
            prev.next = curr.next # skip the current node, by pointing prev's next to curr's next
        else:
            prev = curr # if current node's data is not equal to key, move prev to curr
        curr = curr.next # move curr to curr's next
    return dummy.next # return dummy node's next

# Example usage
# Define the linked list
head = Node(1)
head.next = Node(2)
head.next.next = Node(1)
head.next.next.next = Node(3)

# Call the delete_node function
new_head = delete_node(head, 1)

# Print the linked list after removing all nodes with value 1
print("HEAD->", end="")
curr = new_head
while curr:
    print(curr.data, end="->")
    curr = curr.next
print("NULL")

  
 
```

```python
#Answer of 3
# Define a Node class to create a node of linked list
class LinkedNode:
    def __init__(self, val=0, nxt=None): # initialize value and next for node
        self.val = val
        self.next = nxt

# Define a function to delete all duplicated value elements in a linked list
def remove_duplicate_values(head):
    dummy = LinkedNode(0) # create a dummy node with value set to 0 and point to head
    dummy.next = head # dummy node's next points to head
    curr = dummy # initialize curr as dummy node
    while curr.next: # loop through linked list till the end
        runner = curr # initialize runner as curr
        while runner.next: # loop through linked list from runner till the end
            if runner.next.val == curr.val: # if runner's next node's value is equal to current node's value
                runner.next = runner.next.next # skip the runner's next node, by pointing runner's next to runner's next's next
            else:
                runner = runner.next # if runner's next node's value is not equal to current node's value, move runner to runner's next
        curr = curr.next # move curr to curr's next
    return dummy.next # return dummy node's next

# Example usage
# Define the linked list
head = LinkedNode(1)
head.next = LinkedNode(2)
head.next.next = LinkedNode(1)
head.next.next.next = LinkedNode(3)
head.next.next.next.next = LinkedNode(2)

# Call the remove_duplicate_values function
new_head = remove_duplicate_values(head)

# Print the linked list after removing all duplicated value elements
print("HEAD->", end="")
curr = new_head
while curr:
    print(curr.val, end="->")
    curr = curr.next
print("NULL")
```

```python
#Answer of 4
# Node class definition
class Node:
    # Initialize the node
    def __init__(self, data):
        self.data = data  # store the data value
        self.next = None  # initialize the next node value as None

# Linked List class definition
class LinkedList:
    # Initialize the head node
    def __init__(self):
        self.head = None  # initialize the head node value as None

    # Function to print the linked list
    def printList(self):
        # Store the current node
        current_node = self.head
        current_node.data = "Head->"
        # Loop through all the nodes and print the data
        while current_node:
            print(current_node.data, end="->")
            current_node = current_node.next
        print("NULL")

    # Function to reverse the linked list
    def reverse(self):
        # Initialize the previous node and current node as None
        prev_node = None
        current_node = self.head
        # Loop through all the nodes
        while current_node is not None:
            # Store the next node
            next_node = current_node.next
            # Update the next node of the current node as the previous node
            current_node.next = prev_node
            # Update the previous node as the current node
            prev_node = current_node
            # Update the current node as the next node
            current_node = next_node
        # Update the head node as the previous node
        self.head = prev_node

# Main function definition
if __name__ == '__main__':
    # Create the linked list
    ll = LinkedList()
    # Add nodes to the linked list
    ll.head = Node(1)
    node2 = Node(2)
    node3 = Node(3)
    ll.head.next = node2
    node2.next = node3
    # Reverse the linked list
    ll.reverse()
    # Print the linked list
    ll.printList()
```

```python
#Answer of 5 
# Definition for singly-linked list
class ListNode:
    # Constructor to initialize the node object
    def __init__(self, val=0, next=None):
        self.val = val # assign the value to node
        self.next = next # initialize the next value to None

# Function to calculate average of the list
def Average_List(head): 
    # Check if the list is empty
    if head is None: 
        return 0
    # Point to the head of the list
    node = head 
    total = 0 # Initialize the total to 0
    count = 0 # Initialize the count to 0
    # Traverse the list
    while (node != None): 
        # Add the node value to the total
        total += node.val 
        # Move to the next node
        node = node.next
        # Increase the count by 1
        count += 1 
    # Return the average
    return total/count 
  
# Driver code 
# Initialize the linked list
head = ListNode(1) 
nodeB = ListNode(2) 
nodeC = ListNode(3) 
  
# Connect the nodes
head.next = nodeB 
nodeB.next = nodeC 
  
# Calculate and print the average
print("Average of the list is:", Average_List(head)) 
```
